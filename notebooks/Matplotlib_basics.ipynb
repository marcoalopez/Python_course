# Matplotlib: the Python (default) plotting library

> **What it is?**  
Matplotlib is the standard data visualization tool in Python. It allows creating static, animated or interactive visualizations. Many other Python visualization tools (e.g. seaborn) are built on top of Matplotlib. See https://matplotlib.org/   
>
> **What it is used for?**  
Analize, explore, and show relationships between data.

```python
# to make plots you would normally import the pyplot submodule as follows
import matplotlib.pyplot as plt

# you may also import matplotlib, though this is normally used to change library defaults, etc.
import matplotlib as mpl
```

```python
# make a simple plot using the MATLAB-like interface (normally used for one-liners)
a = [1, 2, 3, 4]
b = [1, 4, 9, 16]

plt.plot(a, b)
```

> Depending on the IDE you are using, you may need to use the ``plt.show()`` command to display the figures. This is not necessary here since by default Jupyter notebooks use waht is called the "inline backend" which means that the plots are shown automatically within the document.

### Customising the plots

you can customimse the plot further by adding the type of marker, its size, etac. or adding labels and titles (almost anything you can think of really). For example

```python
plt.plot(a, b, marker='o', color='orangered', markersize=12)
plt.xlabel('this is label x', fontsize=20)
plt.ylabel('here we use Latex: $x^2$', fontsize=18)
```

```python
# by default, when you call the plot() funtion after customising labels, titles, etc. this will create a new plot
plt.plot(a, [2, 4, 6, 8], marker='o', color='green', markersize=12)
```

```python
# if you call the plot() function before this will plot the data on the same plot
plt.plot(a, b, marker='o', markersize=12, label='$x^2$')
plt.plot(a, [2, 4, 6, 8], 'v--', markersize=12, label='$2x$')

plt.grid(True)
plt.xlabel('this is label x', fontsize=20)
plt.legend(fontsize=18)
```
