# Matplotlib: the Python (default) plotting library

> **What it is?**  
Matplotlib is the standard data visualization tool in Python. It allows creating static, animated or interactive visualizations. Many other data visualization tools for Python (e.g. seaborn) are built on Matplotlib. See https://matplotlib.org/   
>
> **What it is used for?**  
Analize, explore, and show relationships between data.
